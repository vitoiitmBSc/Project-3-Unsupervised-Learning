```python
# Import all the given datasets and explore shape and size
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas_profiling
```

```python
# Import all the given datasets and explore shape and size
company=pd.read_excel('Part2 - Company.xlsx')

company.shape
company.head(40)
```

```python
company_class= company.groupby(['Quality'])
company_class.count()
```

```python
# SO AS WE CAN SEE FROM THIS DATA, THERE ARE 2 QUALITIES THAT ARE AVAILABLEIN THE
# COMPANY QUALITY A AND QUALITY B. SO OUT OF 61 DATA POINTS (26+17=43) 43 DATAPOINTS
# HAVE THE QUALITY AND THE REMAINING 18 POINTS ARE MISSING THE QUALITY
# THE STRATEGY EMPLOYED WILL BE ABLE TO BUILD A CLASSIFICATION ALGORITHM LIKE 
# SVM OR KNN OR LOGISTIC TO BE ABLE TO DECIDE THE QUALITY OF THE WINE
```

```python
nan_values=company[company['Quality'].isnull()]
print (nan_values)
nan_values.shape
nan_values_X=nan_values.drop(['Quality'],axis=1)
print(nan_values_X)
```

```python
company_data=company.drop(company.index[[2,5,7,9,14,18,23,27,29,32,35,40,46,52,57,58,59,60]])
```

```python
company_data.head(57)
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
#Assigning numerical values and storing in another column
company_data['Quality_Cat']=labelencoder.fit_transform(company_data['Quality'])
company_data_X=company_data.drop(['Quality','Quality_Cat'],axis=1)
company_data_Y=company_data['Quality_Cat']
company_data_Y
```

```python
# Let us use an SVM model to build a model from the data which has output already
# and then we will use it to predict the output in the other dataset.
# Let us first scale the data
from sklearn.model_selection import train_test_split
company_data_X_train, company_data_X_test, y_train,y_test= train_test_split(company_data_X,company_data_Y,test_size=0.3,random_state=1)

from scipy.stats import zscore

scaled_company_data_X_train=company_data_X_train.apply(zscore)
scaled_company_data_X_test=company_data_X_test.apply(zscore)
scaled_company_data_X_train.head()
```

```python
from sklearn import svm
from sklearn import metrics
SVM_Model = svm.SVC(C=3, kernel='rbf',probability=True)
SVM_Model.fit(scaled_company_data_X_train,y_train)
y_predict_SVM=SVM_Model.predict(scaled_company_data_X_test)
print(metrics.classification_report(y_test,y_predict_SVM))
print(metrics.confusion_matrix(y_test,y_predict_SVM))
y_predict_SVM
```

```python
scaled_company_data_X_test
```

```python
# Use this model to predict the result for the rows where the Quality is not
# filled
from scipy.stats import zscore
scaled_nan_values_X=nan_values_X.apply(zscore)
y_predict_SVM_final=SVM_Model.predict(scaled_nan_values_X)
y_predict_SVM_final
```

```python
# Model -1 Logistic Regression
# Let us fit the scaled data with these new models and check the accuracy
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="liblinear")
model.fit(scaled_company_data_X_train,y_train)

#predict on the Test Data
y_predict_Logistic = model.predict(scaled_company_data_X_test)
print(metrics.classification_report(y_test,y_predict_Logistic))
print(metrics.confusion_matrix(y_test,y_predict_Logistic))
y_predict_Logistic
```

```python
# Use this model to predict the result for the rows where the Quality is not
# filled
from scipy.stats import zscore
scaled_nan_values_X=nan_values_X.apply(zscore)
y_predict_Logistic_final=model.predict(scaled_nan_values_X)
y_predict_Logistic_final
```
