```python
# Correlation heat map
def plot_corr(df, size=20):
    corr=df.corr()
    fig, ax= plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)),corr.columns)
    plt.yticks(range(len(corr.columns)),corr.columns)
```

```python
# Let us scale the data

from scipy.stats import zscore
scaled_IPL_data_X=IPL_data_X.apply(zscore)
scaled_IPL_data_X.shape
```

```python
# USE K- MEANS TO FIND OUT THE OPTIMAL NUMBER OF CLUSTERS IN THE DATA
# TO find the optimal number of clusters using K-means we will have to use 
# Knee test to see where the distortion decreases the most
from sklearn.cluster import KMeans

#Finding optimal no. of clusters in the Scaled Data
from scipy.spatial.distance import cdist
clusters=range(1,60)
meanDistortions=[]

for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(scaled_IPL_data_X)
    prediction=model.predict(scaled_IPL_data_X)
    meanDistortions.append(sum(np.min(cdist(scaled_IPL_data_X, model.cluster_centers_, 'euclidean'), axis=1)) / scaled_IPL_data_X.shape[0])


plt.plot(clusters, meanDistortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with the Elbow Method')
```

```python
# Since we are supposed to ran all the players, we will have 60 groups for
# 60 different players
# Let us first start with K = 60
final_model_Kmeans=KMeans(60)
final_model_Kmeans.fit(scaled_IPL_data_X)
prediction=final_model_Kmeans.predict(scaled_IPL_data_X)

#Append the prediction 
IPL_data_X["GROUP"] = prediction
scaled_IPL_data_X_60 = scaled_IPL_data_X
scaled_IPL_data_X_60["GROUP"] = prediction
print("Groups Assigned : \n")
IPL_data_X.head(60)
```
